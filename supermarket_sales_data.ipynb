```python
import numpy as np
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/SuperMarket Analysis.csv")
quantity= df["Quantity"].to_numpy()
unit_price = df["Unit price"].to_numpy()

revenue= quantity * unit_price

total_sales = np.sum(revenue)
average_sales = np.mean(revenue)

print("Total Sales Revenue:", total_sales)
print("Average Sales per Order:", average_sales)
first_three_sales = revenue[:3]
print("\nFirst three sales revenues:", first_three_sales)
high_quantity_sales = revenue[quantity > 2]
print("\nHigh quantity sales:", high_quantity_sales)

mean_sales = np.mean(revenue)
median_sales = np.median(revenue)
variance_sales = np.var(revenue)
std_dev_sales = np.std(revenue)

print("\nSales Statistics:")
print("Mean:", mean_sales)
print("Median:", median_sales)
print("Variance:", variance_sales)
print("Standard Deviation:", std_dev_sales)
print("\nRevenue per order:", revenue)
discounted_revenue = revenue * 0.90
print("\nDiscounted Revenue:", discounted_revenue)
print("\nInterpretation:")
print("- Total sales indicate overall business performance.")
print("- Average sales help assess typical order value.")
print("- High variance suggests inconsistent order sizes.")
print("- Standard deviation shows how spread out sales values are.")
print("- High quantity orders significantly impact total revenue.")
```
